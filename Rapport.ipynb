# Rapport 

Les emplois liés à la science des données sont parmi les plus discutés sur le marché ces jours-ci, 
alimentés à la fois par le buzz autour de l’industrie et la demande réelle d’un nombre croissant d’entreprises.
Voici ce qu’une analyse de 664 offres d’emploi nous dit sur ce que les employeurs veulent et ce à quoi les chercheurs d’emploi devraient s’attendre.

## RÉSUMÉ

En un mot, les employeurs recherchent principalement des professionnels expérimentés en science des données qui comprennent également les besoins de l’entreprise et peuvent diriger une équipe. Ces postes de direction s’accompagnent d’un salaire plus élevé, en particulier dans les secteurs bancaire et informatique.


## QUE RECHERCHENT LES EMPLOYEURS 

Recherché : Data Scientist, Sunior Data Scientist,Data Engineer¶
Ce sont les trois titres d’emploi les plus fréquemment vus parmi les affichages . Bien que cela reflète la nature de la demande des employeurs, ils ne représentent qu’environ 10 % des affichages. Cela est principalement dû à la myriade de façons désordonnées et myriades dont les titres d’emploi sont utilisés pour décrire le même rôle dans l’industrie.

## Les stages et les modalités de travail temporaire sont rares

On s’attend à ce que les professionnels des données travaillent avec des quantités substantielles de données sensibles. C’est l’une des raisons possibles pour lesquelles la grande majorité des offres d’emploi sont pour des postes permanents, avec très peu de possibilités de stages ou d’emplois temporaires. Mais il se pourrait bien que les entreprises ne fassent pas autant de publicité officielle pour ces rôles.
